```python
# Importing the python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel('FlightPrediction_Train.xlsx')
test_df=pd.read_excel('FlightPrediction_Test.xlsx')
```

```python
final_df=pd.concat([train_df,test_df], ignore_index=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

## Feature Engineering

```python
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_Time']= final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Time'].apply(lambda x : x.split(':')[0])
final_df['Arrival_Min']=final_df['Arrival_Time'].apply(lambda x : x.split(':')[1])
final_df['Arrival_Hour']=final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Min']=final_df['Arrival_Min'].astype(int)
```

```python
final_df.head(1)
```

```python
final_df['Dep_Hour']=final_df['Dep_Time'].apply(lambda x : x.split(':')[0])
final_df['Dep_Min']=final_df['Dep_Time'].apply(lambda x : x.split(':')[1])
final_df['Dep_Hour']=final_df['Dep_Hour'].astype(int)
final_df['Dep_Min']=final_df['Dep_Min'].astype(int)
```

```python
final_df.head(1)
```

```python
final_df.drop(['Date_of_Journey', 'Dep_Time','Arrival_Time'], axis=1, inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0, '2 stops':2,'1 stop':1, '3 stops':3,'4 stops':4,'NaN':2})
```

```python
final_df.drop(['Route'], axis=1, inplace=True)
```

```python
final_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
final_df['Source'] = encoder.fit_transform(final_df['Source'])
final_df['Destination'] = encoder.fit_transform(final_df['Destination'])
final_df['Airline'] = encoder.fit_transform(final_df['Airline'])
final_df['Additional_Info'] = encoder.fit_transform(final_df['Additional_Info'])
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df.drop(6474, axis=0, inplace=True)
final_df.drop(2660, axis=0, inplace=True)
```

```python
final_df['Duration_Hour']= (final_df['Duration'].apply(lambda x : x.split(' ')[0]).str.split('h').str[0])
```
